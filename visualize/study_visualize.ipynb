```python
TITANIC_URL = (
  'https://raw.githubusercontent.com/'
  'datasciencedojo/datasets/master/titanic.csv'
)
titanic = pd.read_csv(TITANIC_URL)
titanic.shape   # (891, 12)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(titanic['Age'], bins=50)
ax.set_xlabel('나이')
ax.set_ylabel('승객수')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(titanic['Age'],
    bins=25,
    color='steelblue',
    edgecolor='white',
    alpha=0.8)
ax.set_xlabel('나이')
ax.set_ylabel('승객 수')
ax.set_title('타이타닉 승객 나이 분포')
plt.tight_layout()
plt.show()
```

```python
surv_class = titanic.groupby('Pclass')['Survived'].sum()
fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(surv_class.index, surv_class.values,
    color=['#2563eb','#ea580c','#16a34a'],
    edgecolor='white')
ax.set_title('등급별 생존자 수')
ax.set_xticks([1,2,3])
ax.set_xticklabels(['1등급','2등급','3등급'])
plt.tight_layout(); plt.show()
```

```python
colors = titanic['Survived'].map(
    {0: 'gray', 1: 'steelblue'})
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(titanic['Age'],
    titanic['Fare'],
    c=colors, alpha=0.4, s=30)
ax.set_title('나이 vs 요금')
plt.tight_layout(); plt.show()
```

# 산점도. 상관관계를 나타냄
A와 B의 관계
A가 Age, B가 Fare, 각각의 특성이 무엇인지 확인하고 그에대한 상관관계를 확인할 수 있음


```python
data = [titanic[titanic['Pclass']==c]['Fare'].dropna()
    for c in [1,2,3]]
fig, ax = plt.subplots(figsize=(7, 5))
ax.boxplot(data,
    tick_labels=['1등급','2등급','3등급'])
ax.set_ylabel('요금')
ax.set_title('등급별 요금 분포')
plt.tight_layout(); plt.show()
```

선을 수염이라고 합니다. 상한, 하한이 있는데, 이걸 계산하는 건, 기본적으로4분위수라는 통게깝싱다. 중앙값 50%, Q1, 25% Q3 75% . 상한 하한은 뭘까?

밖에 있는 것들을 Outlier 이상치라고 한다.
이상치는 데이터에 어떤 영향을 미치느가?
평균이 왜곡이된다.


```python
# 파이 차트
class_counts = titanic['Pclass'].value_counts().sort_index()
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(class_counts.values,
    labels=['1등급','2등급','3등급'],
    autopct='%1.2f%%', # 소수점
    startangle=0, # 각도 시작지점
    wedgeprops=dict(edgecolor='black')) # 경계선 구분 색
ax.set_title('객실 등급별 승객 비율')
plt.tight_layout();
plt.show()

plt.savefig('chart.png',
    dpi=150,
    bbox_inches='tight')
```

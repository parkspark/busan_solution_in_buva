```python
import pandas as pd
import numpy as np
TITANIC_URL = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(TITANIC_URL)
```

```python
# 1. 행 수, 열 수 (rows, columns) 출력
print(f'titanic 데이터셋은 {titanic.shape[0]}행 {titanic.shape[1]}열으로 구성되어 있다.')
```

```python
# 2. 컬럼별 타입, 결측치 확인
titanic.info()
```

info() 함수로 DF의 cloumn 별 정보를 확인할 수 있다.  
Age, Cabin, Embarked 에 결측치가 존재한다.

```python
# 3. 생존자(survived = 1)만 필터링
print(titanic[titanic['Survived'] == 1].shape)
titanic[titanic['Survived'] == 1].head()
```

```python
# 4. 성별에 따른 생존률 계산
print(titanic[(titanic['Survived'] == 1) & (titanic['Sex'] == 'female')].shape)
titanic[(titanic['Survived'] == 1) & (titanic['Sex'] == 'female')].head()
```

```python
# 5. Age 결측치를 중앙값으로 채우기
titanic.isnull().sum() # 어떤 열에서 결측치가 몇개 있는 지 확인
titanic['Age'].fillna(titanic['Age'].median()) #fillna 로 결측치를 다른 값으로 채우기
```

```python
# 6. 등급별 평균 나이, 요금, 생존율(agg 사용)
titanic.groupby('Pclass')
'''위를 실행하면 
<pandas.api.typing.DataFrameGroupBy object at 0x000002A5CB007C50> 
이런 값이 뜨는데 왜 이러는가?'''

titanic.groupby('Pclass').agg(나이 = ('Age','mean'), 요금 = ('Fare','mean'),생존율 = ('Survived','mean'))
```

```python
#7. 1등급 여성 승객 이름 목록
titanic['Name'][titanic['Pclass'] == 1]
```

```python
#8. 35세 초과 + 1등급 + 남성 승객 수
titanic[(titanic['Age'] > 35) & (titanic['Pclass'] == 1) & (titanic['Sex'] == 'male')].shape[0]
```

```python
#9. 탑승항(embark_town)별 생존율 비교
titanic.groupby('Embarked').agg( 생존율 =('Survived', 'mean'))
```

## Chanllenge 쇼핑몰 데이터 전처리

```python
import pandas as pd
import numpy as np
np.random.seed(42)
n = 150
shop = pd.DataFrame({
    'customer': np.random.choice(['김민준','이서현','박지우','최하은','정도윤'], n),
    'category': np.random.choice(['의류','전자기기','식품','생활용품','뷰티'], n),
    'quantity': np.random.randint(1, 6, n),
    'price':    np.random.choice([15000,29000,45000,89000,120000], n),
    'rating':   np.random.choice([1,2,3,4,5], n, p=[0.05,0.10,0.20,0.35,0.30]),
    'returned': np.random.choice([True, False], n, p=[0.1, 0.9])
})
shop.loc[np.random.choice(shop.index, 8, replace=False), 'rating'] = float('nan')
shop.to_csv('shop_orders.csv', index=False)

df_shop = pd.read_csv('shop_orders.csv')
# Q1. 카테고리별 총 매출 내림차순
# Q2. 반품률 가장 높은 카테고리
# Q3. rating 결측치를 category별 평균으로 채우기
# Q4. 고객별 총 주문 금액 TOP 3
```

```python
# Q1. 카테고리별 총 매출 내림차순

# df_shop.groupby('category')
# df_shop.price * df_shop.quantity
df_shop['sales'] = df_shop.price * df_shop.quantity
df_shop.groupby('category').agg( 매출 = ('sales', 'sum')).sort_values(by='매출', ascending=False).reset_index()
```

```python
# Q2. 반품률 가장 높은 카테고리
# df_shop.groupby('category').agg(반품률 = ('returned', 'mean')) # 뷰티가 0 나와서 확인해보니 진짜 0임

df_shop_returned = df_shop.groupby('category').agg(반품률 = ('returned', 'mean')).sort_values(by='반품률',ascending=False).reset_index()
print(df_shop_returned)

df_shop_returned.loc[0,'category']
```

```python
#Q3. rating 결측치를 category별 평균으로 채우기
print(df_shop.isnull().sum())

# df_shop.groupby('category').agg(rating =('rating', 'mean')) # 카테고리별 평균 구하기
# df_shop.groupby('category')['rating'].transform('mean') 
# # transform : 원본 데이터프레임의 행 개수와 동일하게 유지하여 반환합니다. 
# 즉, 각 행의 카테고리에 맞는 평균값이 쭉 나열된 시리즈(Series)를 만듭니다.
df_shop_rating_fillna = df_shop['rating'].fillna(df_shop.groupby('category')['rating'].transform('mean'))

print(df_shop_rating_fillna.isnull().sum())
```

```python
# Q4. 고객별 총 주문 금액 TOP 3

df_shop.groupby('customer').agg(총주문금액 = ('sales', 'sum')).sort_values(by='총주문금액', ascending=False).reset_index().head(3)
```
